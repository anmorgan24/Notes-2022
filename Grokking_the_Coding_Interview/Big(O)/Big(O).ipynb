### Computational Complexities
- Clock time is hardware-dependent
- An efficient program may take more time to execute on a slower computer than an inefficient program on a fast computer
- So, clock time is not a good metric to find time complexity
- Big(O) allows us to compare the algorithm complexity of two programs in a standardized manner
- **Big(O) notation is an algorithm complexity metric that defines the relationship between the number of inputs and the steps taken by the program to proccess those inputs.**
- Big(O) is **not** about measuring speed
- Big(O) **is** about measuring the amount of work a program has to do *as an input scales.*
- We can use Big(O) to define **both** time and space complexity
- Computational complexity is generally expressed as a function of `n`, where **`n` is the size of the input.**
- Big-O notation, sometimes called **asymptotic notation**, is a mathematical notation that describes the limiting behavior of a function when the argument tends towards a particular value or infinity.
- **Asymptotics** is a method of describing limiting behavior. For example, suppose we are interested in the properties of a function `f` as `n` (inputs) becomes very large.
- Big-O notation characterizes functions according to their growth rates: different functions with the same growth rate may be represented using the same O notation.

## CONSTANT: 
Steps taken remain the same irrespective of the input size; does **not** scale with input.

An algorithm is said to have a constant time when it is not dependent on the input data (n). **No matter the size of the input data, the running time will always be the same.** 

$$O(c)$$
<img src='data/bigo1.png' width="400" height="200" align="center"/>

```python
def display_first_cube(items):
    result = pow(items[0],3)
    print (result)

inputs = [2, 3, 4, 5, 6, 7]
display_first_cube(inputs)
```

```python
num_of_inputs = np.array([1,2,3,4,5,6,7])
steps = [2 for n in num_of_inputs]
 
plt.plot(num_of_inputs, steps, 'r')
plt.xlabel('Number of inputs')
plt.ylabel('Number of steps')
plt.title('O(c) Complexity')

plt.show()
```

***
## LOGARITHMIC:

An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step (it doesnâ€™t need to look at all values of the input data). Algorithms with logarithmic time complexity are commonly found in operations on binary trees or when using binary search.

It is important to understand that an algorithm that must access all elements of its input data cannot take logarithmic time, as the time taken for reading input of size n is of the order of n.



$$O(log(n))$$




```python
def binary_search(data, value):
    n = len(data)
    left = 0
    right = n - 1
    while left <= right:
        middle = (left + right) // 2
        if value < data[middle]:
            right = middle - 1
        elif value > data[middle]:
            left = middle + 1
        else:
            return middle
    raise ValueError('Value is not in the list')
    
if __name__ == '__main__':
    data = [1, 2, 3, 4, 5, 6, 7, 8, 9]
    print(binary_search(data, 8))
```

```python
num_of_inputs = np.array([1,2,3,4,5,6,7])
steps = [np.log(n) for n in num_of_inputs]

plt.plot(num_of_inputs, steps, 'r')
plt.xlabel('Number of inputs')
plt.ylabel('Number of steps')
plt.title('O(n) Complexity: Linear')
plt.show()
```

***
## LINEAR:
Here, a single unit increase in the input causes a unit increase in the steps required to complete the program execution. The example below, which calculates the cubes of all elements in a list, has linear complexity. This is because as the input grows, it will need to do one more unit of work per item. If you double the elements in the input list, the steps needed to execute the function will also be doubled. Functions with linear complexity should have a straight line increasing in a positive direction.

$$O(n)$$
<img src='data/bigo2.png' width="400" height="200" align="center"/>

```python
def display_all_cubes(items):
    for item in items:
        result = pow(item,3)
        print (result)

inputs = [2,3,4,5,6,7]
display_all_cubes(inputs)
```

```python
num_of_inputs = np.array([1,2,3,4,5,6,7])
steps = [2*n for n in num_of_inputs]

plt.plot(num_of_inputs, steps, 'r')
plt.xlabel('Number of inputs')
plt.ylabel('Number of steps')
plt.title('O(n) Complexity: Linear')
plt.show()
```

***
## QUADRATIC:

In a function with quadratic complexity, the output steps increase quadraticlly with the increase in inputs. In the example below, the function multiplies each item in the "items" list with each of the other elements. The outer loop iterates through each item, and the for each item in the outer loop, the inner loop iterates over each item. This makes the total number of steps `n x n`, where `n` is the number of items in the "items" list

**$$O(n^2)$$**

<img src='data/bigo3.png' width="500" height="250" align="center"/>

```python
num_of_inputs = np.array([1,2,3,4,5,6,7])
steps = [pow(n,2) for n in num_of_inputs]

plt.plot(num_of_inputs, steps, 'r')
plt.xlabel('Number of inputs')
plt.ylabel('Number of steps')
plt.title('O(n^2) Complexity: Quadratic')
plt.show()
```

### CUBIC:

$$O(n^3)$$



```python
num_of_inputs = np.array([1,2,3,4,5,6,7])
steps = [pow(n,3) for n in num_of_inputs]

plt.plot(num_of_inputs, steps, 'r')
plt.xlabel('Number of inputs')
plt.ylabel('Number of steps')
plt.title('O(n^3) Complexity: Cubic')
plt.show()
```

***
### EXPONENTIAL:
$$O(2^n)$$

```python
num_of_inputs = np.array([1,2,3,4,5,6,7])
steps = [pow(2,n) for n in num_of_inputs]

plt.plot(num_of_inputs, steps, 'r')
plt.xlabel('Number of inputs')
plt.ylabel('Number of steps')
plt.title('O(2^n) Complexity: Exponential')
plt.show()
```

***
### FACTORIAL:
$$O(n!)$$

```python
from math import factorial
```

```python
num_of_inputs = np.array([1,2,3,4,5,6,7])
steps = [factorial(n) for n in num_of_inputs]

plt.plot(num_of_inputs, steps, 'r')
plt.xlabel('Number of inputs')
plt.ylabel('Number of steps')
plt.title('O(n!) Complexity: Factorial')
plt.show()
```

#### Important Note:
It is important to note that when analyzing the time complexity of an algorithm with several operations we need to describe the algorithm based on the largest complexity among all operations. For example:

```python
def my_function(data):
    first_element = data[0]
    
    for value in data:
        print(value)
    
    for x in data:
        for y in data:
            print(x, y)
```

This function, though it doesn't make much sense, clearly has multiple time complexities: $O(1)$ + $O(n)$ + $O(nÂ²)$. So, when increasing the size of the input data, the bottleneck of this algorithm will be the operation that takes $O(nÂ²)$. Based on this, we can describe the time complexity of this algorithm as $O(nÂ²)$.

## Big-Theta ($\theta$-notation)
- Big O notation provides an upper bound to a function whereas Big Theta provides a tight bound.
- Theta notation encloses the function from above and below. Since it represents the upper and the lower bound of the running time of an algorithm, it is used for analyzing the average-case complexity of an algorithm. Theta bounds the function within constants factors.
